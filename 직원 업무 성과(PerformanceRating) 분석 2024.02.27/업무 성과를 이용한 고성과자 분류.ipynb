```python
# 나이대 범주 생성 (18살부터 시작하여 5살 간격)
bins = list(range(18, data['Age'].max() + 5, 5))
labels = [f'{i}-{i+4}' for i in bins[:-1]]  # 예: '18-22', '23-27', ...
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# 나이대별 시급 분포 시각화
plt.figure(figsize=(12, 8))
sns.boxplot(x='AgeGroup', y='HourlyRate', data=data)
plt.title('Hourly Rate Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Hourly Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
top_30_percent_by_age_group = data.groupby('AgeGroup').apply(lambda x: x.sort_values(by='HourlyRate', ascending=False).head(int(len(x) * 0.35)))
pd.set_option('display.max_columns', None)  # 모든 열을 출력하도록 설정
# 결과 출력 (인덱스 재설정)
top_30_percent_by_age_group = top_30_percent_by_age_group.reset_index(drop=True)
top_30_percent_by_age_group 
```

```python
# 각 나이대별로 시급이 하위 65%에 해당하는 데이터 선택
bottom_65_percent_by_age_group = data.groupby('AgeGroup').apply(lambda x: x[x['HourlyRate'] <= x['HourlyRate'].quantile(0.65)])

# 결과 출력 (인덱스 재설정)
bottom_65_percent_by_age_group = bottom_65_percent_by_age_group.reset_index(drop=True)
pd.set_option('display.max_columns', None)

# 하위 65% 데이터 출력
bottom_65_percent_by_age_group
```

```python
'''순서가 중요한 범주변수는 LabelEncoding, 순서가 중요하지 않은 범주변수는 OneHotEncoding 처리
LabelEncoding =  Education, JobLevel
OneHotEncoding = Attrition, Department, BusinessTravel,EducationField, Gender, JobRole, MaritalStatus
OverTime변수는 이따 상관분석을 해야하기때문데 Yes:1, No:0으로 변환'''
from sklearn.preprocessing import LabelEncoder

Real_high_performer['OverTime'] = Real_high_performer['OverTime'].replace({'Yes': 1, 'No': 0})
Real_low_performer['OverTime'] = Real_low_performer['OverTime'].replace({'Yes': 1, 'No': 0})

Le = LabelEncoder()
columns_to_encode = ['Education', 'JobLevel']
for column in columns_to_encode:
    Real_high_performer[column] = Le.fit_transform(Real_high_performer[column])
    Real_low_performer[column] = Le.fit_transform(Real_low_performer[column])
    
Real_high_performer= pd.get_dummies(Real_high_performer, columns = ['Attrition','Department','BusinessTravel','EducationField','Gender','JobRole','MaritalStatus'])
Real_low_performer= pd.get_dummies(Real_low_performer, columns = ['Attrition','Department','BusinessTravel','EducationField','Gender','JobRole','MaritalStatus'])
```

```python
corr_matrix = Real_high_performer.corr()

# 히트맵으로 상관관계 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.show()
```

```python
# 상관관계 행렬 계산
corr_matrix_TWY = Real_high_performer.corr().abs()  # 절대값을 사용하여 양의 상관관계와 음의 상관관계를 동일하게 처리

# 상관관계 행렬의 상단 삼각형을 NaN으로 설정
upper = corr_matrix.where(pd.np.triu(pd.np.ones(corr_matrix.shape), k=1).astype(pd.np.bool))

# 상관관계가 높은 상위 10개의 쌍을 찾음
top_correlations_TWY = upper.unstack().sort_values(ascending=False).dropna()

# 상위 10개 쌍 출력
print(top_correlations_TWY.head(10))
```

```python
# 'TotalWorkingYears'와 다른 모든 변수 간의 상관관계 계산
correlations_TWY = Real_high_performer.corr()['TotalWorkingYears']

# 상관관계가 0.5 이상 또는 -0.5 이하인 변수들을 필터링
significant_correlations_TWY = correlations_TWY[(correlations_TWY >= 0.5) | (correlations_TWY <= -0.5)]

# 결과 출력
print("TotalWorkingYears와 상관관계가 0.5 이상이거나 -0.5 이하인 변수들:")
print(significant_correlations_TWY)
```
